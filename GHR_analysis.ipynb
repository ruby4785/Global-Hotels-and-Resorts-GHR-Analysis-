```python
import calendar

def get_weeks_in_month(year, month):
    num_days = calendar.monthrange(year, month)[1]
    first_day_of_month = calendar.weekday(year, month, 1)
    num_weeks = (num_days + first_day_of_month) // 7
    if (num_days + first_day_of_month) % 7 > 0:
        num_weeks += 1
    return num_weeks


year = 2016
for month in range(1, 13):
    num_weeks = get_weeks_in_month(year, month)
    print(f"Number of weeks in {calendar.month_name[month]} {year}: {num_weeks}")
```
